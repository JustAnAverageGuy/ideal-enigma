```python
import pandas as pd
import numpy as np

import requests
BASE_URL = r'https://codeforces.com/api/'
METHOD = r'user.ratedList'
r = requests.get(
    BASE_URL + '/' + METHOD,
    {
        # "handles": ";".join(s),
        "contestId": "1849",
    }
)

print(r.status_code)

s = r.json()

users = s['result']
```

```python
rated_users = [user for user in users if "rating" in user]

df = pd.DataFrame(rated_users)
# df[df["rating"] != np.NaN].sort_values("rating",ascending=False)[["handle","rating"]]

# len(rated_users)
df[(100 < df['rating']) & (df["rating"] < 1800)][["handle","rating"]]
```

```python
class User:
    def __init__(self, **kwags):
        self.data = {'country': None,
                     'lastOnlineTimeSeconds': None,
                     'city': None,
                     'rating': None,
                     'friendOfCount': None,
                     'titlePhoto': None,
                     'handle': None,
                     'avatar': None,
                     'firstName': None,
                     'contribution': None,
                     'organization': None,
                     'rank': None,
                     'maxRating': None,
                     'registrationTimeSeconds': None,
                     'maxRank': None}
        for i, j in kwags:
            if i not in self.data:
                raise KeyError(f"Extra property {i} not found")
            self.data[i] = j
```

```python
usernames = [user["handle"] for user in users if (user["lastOnlineTimeSeconds"] - 1690468200) > 0]
```
